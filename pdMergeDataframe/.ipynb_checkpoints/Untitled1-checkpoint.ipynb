```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
data = pd.read_csv('content/Product_data_cmGokul1.csv')
```

```python
data.head()
```

```python
flitered_data  = data[(data['FAT'] == 3.5) & (data['SNF'] == 7.5)]
```

```python
print(flitered_data)
```

```python
# Convert RateDate to datetime
flitered_data['RateDate'] = pd.to_datetime(flitered_data['RateDate'])
```

```python
# Plot time series for each variable
flitered_data.plot(x='RateDate', y=['FAT', 'SNF', 'Rate'])
```

```python
# Check for stationarity using Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller
```

```python
def is_stationary(series):
  result = adfuller(series)
  return result[1] < 0.05

is_stationary(flitered_data['Rate'])
# Perform similar checks for SNF and Rate
```

```python
# Split data into train and test sets
train_size = int(len(flitered_data) * 0.8)
train, test = flitered_data[:train_size], flitered_data[train_size:]
```

```python
# Define and train ARIMA model with initial hyperparameters
model = ARIMA(train['Rate'], order=(1, 1, 1))
model_fit = model.fit()
```

```python
# Predict values for the test set
predictions = model_fit.forecast(steps=len(test))
```

```python
# Plot training data, test data, and predictions
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Rate'], label='Training Data')
plt.plot(test.index, test['Rate'], label='Test Data')
plt.plot(test.index, predictions, label='Predictions')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('ARIMA Model - Training, Testing, and Predictions')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Print prediction values
print("Predicted Rates:")
for i, prediction in enumerate(predictions):
    print(f"Date: {test.iloc[i]['RateDate']}, Predicted Rate: {prediction:.2f}")
```
